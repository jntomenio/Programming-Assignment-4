### 1. Data Filtering

```python
# Imports the pandas library with an alias pd
import pandas as pd

board = pd.read_excel('board2.xlsx')

board
```

```python
# To display a dataframe wherein the Name, GEAS, and Electronics grade more than 70 are displayed
# With people who only lives at Luzon and has a track of Instrumentation. 
#The .loc() function is going to be used, with multiple, and conditions and only a specific columns that are going to be shown

Instru = board.loc[
    (board['Hometown'] == 'Luzon') & 
    (board['Track'] == 'Instrumentation') & 
    (board['Electronics'] > 70), 
    ['Name','GEAS','Electronics'] ]

Instru
```

```python
# To display a dataframe wherein the Name, Track, Electronics, and Average greater or equal to 55 are displayed; With people who only lives at Mindanao and are female. The .loc() function is going to be used, with multiple AND conditions and only a specific columns that are going to be shown. The additional key named "Average" is going to be added with the .mean() function that is going to be utilized. The .mean() function will only compute for the mean of each row (4 subjects) because its set at axis=1 and it will be rounded by 2 using the round() function.
board['Average'] = round(board[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1),2)

Mindy = board.loc[
    (board['Hometown'] == 'Mindanao') & 
    (board['Gender'] == 'Female') & 
    (board['Average'] >= 55), 
    ['Name','Track','Electronics', 'Average'] ]

Mindy
```

 ### 2. Data Visualization

```python
# It is essential to import the matplotlib.pyplot library as it will be required for making different graphs
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,4))
plt.bar(board['Hometown'],board['Average'])
```

- It was shown that Luzon has the highest average grade with Mindanao being the lowest.

```python
plt.figure(figsize=(10,4))
plt.bar(board['Track'],board['Average'])
```

- It was shown that Microelectronics has the highest average grade, with Instrumentation being the lowest.

```python
plt.figure(figsize=(10,4))
plt.bar(board['Gender'],board['Average'])
```

- It was shown that females perform well and have higher average grades than males.
